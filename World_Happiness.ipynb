```python
#global import
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data2015 = pd.read_csv("Dataset/2015.csv")
data2016 = pd.read_csv("Dataset/2016.csv")
data2017 = pd.read_csv("Dataset/2017.csv") #No Region column; Different column name format -> Add column Region, compare with data2015+2016
data2018 = pd.read_csv("Dataset/2018.csv") #Country or Region column mixed
data2019 = pd.read_csv("Dataset/2019.csv") #Country or Region column mixed

data2016.head()
```

```python
data2016['Region'].value_counts()
```

```python
#Look up Country in data2017, 2018 and 2019 by Country in 2016

data2017_with_region = data2017.merge(
    right=data2016[['Country', 'Region']],
    how = 'left'
)
data2017_with_region.head()
```

```python

data2018_with_region = data2018.merge(
    right=data2016[['Country', 'Region']],
    how = 'left',
    left_on = 'Country or region',
    right_on = 'Country'
)
data2018_with_region.head()
```

```python
data2019_with_region = data2019.merge(
    right=data2016[['Country', 'Region']],
    how = 'left',
    left_on = 'Country or region',
    right_on = 'Country'
)
data2019_with_region.head()
```

```python
# Find mean/median for each region over the years
# Western Europe region : 2015-2019

happiness_score_2015 = data2015.groupby(['Region'])['Happiness Score'].mean().reset_index()
print(happiness_score_2015)
```

```python
happiness_score_2016 = data2016.groupby(['Region'])['Happiness Score'].mean().reset_index()
print(happiness_score_2016)
```

```python
happiness_score_2017 = data2017_with_region.groupby(['Region'])['Happiness.Score'].mean().reset_index()
print(happiness_score_2017)
```

```python
happiness_score_2018 = data2018_with_region.groupby(['Region'])['Score'].mean().reset_index()
print(happiness_score_2018)
```

```python
happiness_score_2019 = data2019_with_region.groupby(['Region'])['Score'].mean().reset_index()
print(happiness_score_2019)
```

```python

happiness_data = {
   'Region': np.sort(data2015['Region'].unique()),
   '2015': happiness_score_2015['Happiness Score'],
   '2016' : happiness_score_2016['Happiness Score'],
   '2017' : happiness_score_2017['Happiness.Score'],
   '2018' : happiness_score_2018['Score'],
   '2019' : happiness_score_2019['Score'],
}

happiness_score_by_year = pd.DataFrame(happiness_data)
print(happiness_score_by_year)
```
